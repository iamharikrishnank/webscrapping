```python
import requests
from bs4 import BeautifulSoup as BSoup
import pandas as pd
requests.packages.urllib3.disable_warnings()
```

```python
from bs4 import BeautifulSoup as BSoup
import requests
import re
skill_list = []
company_list = []
experience_list = []
skill_list = []
position_list = []

with requests.Session() as s:
    for page in range(1,100):
        headers = {'User-Agent' : 'Mozilla/5.0'}
        url = 'https://www.timesjobs.com/candidate/job-search.html?from=submit&searchType=personalizedSearch&txtLocation=Cochin/%20Kochi/%20Ernakulam&luceneResultSize=25&postWeek=60&pDate=Y&sequence=' + str(page)
#         print(url)
        r = s.get(url, headers = headers)
        soup = BSoup(r.content,'html.parser')
#         print(url)
       
        tables = soup.find_all('li', {"class": "clearfix job-bx wht-shd-bx"})
#         print(tables)
        for table in tables:
           position = table.find('h2').text.split(' ',1)[1]
           company = table.find('h3', {"class": "joblist-comp-name"}).text.split(' ',1)[1].split('\r',1)[0]
           experience = table.find_all('li')[0].text.split('-',1)[1]
           skillset    =  table.find('span', {"class": "srp-skills"}).text.split(' ',1)[1]
           skills  = skillset.split('\n',1)[1]
           position_list.append(position)
           company_list.append(company)
           experience_list.append(experience)
           skill_list.append(skills)
           
       
```

```python
df=pd.DataFrame()
```

```python
df['Job Position']=position_list
df['Company']=company_list
df['Experience']=experience_list
df['Skills']=skill_list
df['City']='Kochi'
```
