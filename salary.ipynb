```python
import requests
from bs4 import BeautifulSoup as BSoup
import pandas as pd
requests.packages.urllib3.disable_warnings()
```

```python
from bs4 import BeautifulSoup as bs
import requests
import re

results = []
job_titles=[]
avgsal_list=[]
company_list=[]
minsal_list=[]
maxsal_list=[]

with requests.Session() as s:
    for page in range(100):
        headers = {'User-Agent' : 'Mozilla/5.0'}
        url = "https://www.glassdoor.co.in/Salaries/cochin-salary-SRCH_IL.0,6_IM1082_IP" + str(page) + ".htm?clickSource=searchBtn'"
        r = s.get(url, headers = headers)
        soup = BSoup(r.content,'html.parser')
#         print(soup)
        News = soup.find_all('div', {"data-test": "salary-list-items"})
        for postions in News:
            pos = postions.find('p', {"class": "m-0"})
            
            
            avg_sal = postions.find('div', {"class": "col-2 d-none d-md-flex flex-row justify-content-end"})
            comp = postions.find_all('p')[1]
            min_sal = postions.find('div', {"class": "col-3 offset-1 d-none d-md-block"}).find_all('span')[0]
            max_sal = postions.find('div', {"class": "col-3 offset-1 d-none d-md-block"}).find_all('span')[1]
            company = comp.get_text()
            positions = pos.get_text()
            try:
                
                avg_salary = avg_sal.text
            except:
                
                avg_salary = ' '
            job_titles.append(positions)
            avgsal_list.append(avg_salary)
            company_list.append(company)
            minsal_list.append(min_sal.text)
            maxsal_list.append(max_sal.text)
            print(avg_salary)
```

```python
df=pd.DataFrame()
```

```python
df['Job Title']=job_titles
df['Company']=company_list
df['Average Salary']=avgsal_list
df['Minimum Salary']=minsal_list
df['Maximum Salary']=maxsal_list
df['City']='Kochi'
```
