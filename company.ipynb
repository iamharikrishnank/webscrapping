```python
import requests
from bs4 import BeautifulSoup 
import pandas as pd
requests.packages.urllib3.disable_warnings()
```

```python
from bs4 import BeautifulSoup as BSoup
import requests
import re

company_list = []
size_list = []
industry_list = []
rate_list = []

with requests.Session() as s:
    for page in range(100):
        headers = {'User-Agent' : 'Mozilla/5.0'}
        url = 'https://www.glassdoor.co.in/Explore/browse-companies.htm?overall_rating_low=3.5&page='+ str(page)+'&isHiringSurge=0&locId=2887994&locType=C&locName=Kochi'
        r = s.get(url, headers = headers)
        soup = BSoup(r.content,'html.parser')
        tables = soup.find_all('div', {"class": "row d-flex flex-wrap"})
        for table in tables:
            companies = table.find_all('h2', {"data-test": "employer-short-name"})
            for company in companies:
                company = company.text
                company_list.append(company)
            company_sizes = table.find_all('span', {"data-test": "employer-size"})
            for size in company_sizes:
                company_size = size.text
                size_list.append(company_size)
            industries = table.find_all('span', {"data-test": "employer-industry"})
            for industry in industries:
                industry = industry.text
                industry_list.append(industry)
            ratings = table.find_all('span', {"data-test": "rating"})
            for rate in ratings:
                rate = rate.text
                rate_list.append(rate)
       
```

```python
df=pd.DataFrame()
```

```python
df['Company']=company_list
df['Company Size']=size_list
df['Industry']=industry_list
df['Ratings']=rate_list
df['City']='Kochi'
```
