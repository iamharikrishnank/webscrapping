```python
import requests
from bs4 import BeautifulSoup 
import pandas as pd
requests.packages.urllib3.disable_warnings()
```

```python
from bs4 import BeautifulSoup as BSoup
import requests
import re
rent_list = []
name_list = []
sqft_list = []
bhk_list = []

city = 'bangalore'
with requests.Session() as s:
    for page in range(100):
        headers = {'User-Agent' : 'Mozilla/5.0'}
        url = 'https://www.99acres.com/property-for-rent-in-'+city+'-ffid-page-'+str(page)
        r = s.get(url, headers = headers)
        soup = BSoup(r.content,'html.parser')
        
        
        tables = soup.find_all('table', {"class": "srpTuple__tupleTable"})
        for table in tables:
           rents_cal =  table.find_all('tr')[2].find_all('td')[0]
           rent = rents_cal.text.split('Deposit',1)[0]
           name =  table.find_all('tr')[1].find_all('td')[0].text
           sqft  =  table.find_all('tr')[2].find_all('td')[1].text.split('(',1)[0]
           bhk  =  table.find_all('tr')[2].find_all('td')[2].text.split('K',1)[0] + 'K'
           rent_list.append(rent)
           name_list.append(name)
           sqft_list.append(sqft)
           bhk_list.append(bhk)
       
```

```python
df=pd.DataFrame()
```

```python
df['Flat']=name_list
df['Rent']=rent_list
df['BHK']=bhk_list
df['SQFT']=sqft_list
df['City']='Banglore'
```
